```python
conn.execute('''CREATE TABLE DATA
    (uniqueID INTEGER PRIMARY KEY,
    drugName CHAR(60) NOT NULL,
    condition CHAR(60) NOT NULL,
    review TEXT NOT NULL,
    rating INTEGER NOT NULL,
    date TEXT NOT NULL,
    usefulCount INTEGER NOT NULL);''')
print("Table created successfully")
```

```python
with open("C:/data/Drug_Clean.csv") as csv_file:
    csv_reader=csv.reader(csv_file,delimiter=',')
    for row in csv_file:
        print(row)
        uniqueID=int(row[0])
        drugName=row[1]
        condition=row[2]
        review=row[3]
        rating=int(row[4])
        date=row[5]
        usefulCount=int(row[6])
        conn.execute('''INSERT INTO DATA(uniqueID,drugName,condition,review,rating,date,usefulCount) VALUES(?,?,?,?,?,?,?)''',(uniqueID,drugName,condition,review,rating,date,usefulCount))
        conn.commit()
```
